```python
from ultralytics import YOLO
model = YOLO("yolov8n-seg.pt")
```

```python
model.train(
    data="/content/dataset/data.yaml",
    epochs=50,
    imgsz=640,
    batch=8
)
```

```python
model.predict(
    source="/content/71Xw+VDIkVL._AC_SL1500_.jpg",
    save=True,
    conf=0.25
)
```
